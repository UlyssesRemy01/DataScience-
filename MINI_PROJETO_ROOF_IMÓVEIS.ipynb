```python
#importando dataframe fornecido pelo cliente.
import pandas as pd 
path = r'/content/kc_house_data.csv'
df = pd.read_csv(
                 filepath_or_buffer = path,
                 sep = ',',
                 decimal = '.',
        )
df
```

```python
#Verificando tipos de variáveis
df.info()
```

```python
df.isna().sum()
```

```python
#Verificando duplicatas
df.duplicated().sum()
```

```python
#corrigindo a coluna Date usando o pd.datetime
df["date"] = pd.to_datetime(
            arg = df['date'],
            format = '%Y%m%dT%H%M%S'
)
df
```

```python
# compreendendo a distribuição dos preços e classificando-os em padrões
# Baixo padrão < 321837;
# 3218375 < Médio < 645000
# 645000 < Alto < 1127000
# 1127000 < Luxo
```

```python
#Criando uma coluna para os padrões de imóvel
df.loc[df['price'] < 321837, 'padrao' ] = 1
df.loc[(df['price'] > 321837) & (df['price'] < 645000), 'padrao' ] = 2
df.loc[(df['price'] > 645000) & (df['price'] < 1127000), 'padrao' ] = 3
df.loc[df['price'] > 1127000, 'padrao' ] = 4
```

```python
#Distribuição dos preços em boxplot de padrões
px.box(df, x='padrao',y='price')
```

```python
#criando a coluna price_ft²
df["price_ft²"] = df['price'] / df['sqft_lot']
df["price_ft²"] = df["price_ft²"].astype(float).round(2) #=> arredondando floats
df
```

```python
#conhecendo e plotando as estatísticas de de 'price_ft²'
px.box(df, x= 'price_ft²')
```

```python
"""criando coluna cidades no dataframe usando biblioteca
 de importação de zipcode com tratamento
 de erro de exceção para zip code faltante. pois na primeira importação 
 foram constatados 198 valores vazios para um determinado zip code"""
# Criando tabela com os nomes da cidades. 
import pyzipcode

zcdb = ZipCodeDatabase()
city = []

for zipcode in df["zipcode"]:

    try:
         zc = zcdb[zipcode]
         city.append(zc.city)

    except KeyError:
        if zipcode == 98077:
           city.append(str("Woodinville"))
        else:
            city.append(str('nan'))

df['city'] = city 
df2 = df
df2
```

```python
#contando quantidade de imóveis por cidade 
dfplot = df2['city'].value_counts()
dfplot
```

```python
# Criando gráfico para comparação e visualização de quais cidades possuem mais imoveis no dataframe
sns.barplot(x=dfplot.index, y=dfplot.values)
plt.title('Registros de transação de vendas de imóveis por cidade no condado de King - WH -EUA')
plt.xlabel('Cidades')
plt.xticks(rotation=90)
plt.ylabel('Transações de Venda')
plt.show()
```

```python
df.columns
```

```python
#criando a coluna para custo benefício
df["custo-benefício"] = (df["sqft_lot"]/df["price"]) * 100
df["custo-benefício"] = df["custo-benefício"].astype(int) # converter para inteiro
df["custo-benefício"] = df["custo-benefício"].apply(lambda x: "{:.2f}%".format(x/100)) # formatar número com porcentagem
df
```

```python
#Criando dataframe para elaboração de produto final selecionando as colunas de df 
df2 = df[['id', 'city', 'price', 'price_ft²', 'bedrooms', 'bathrooms', 
          'sqft_lot', 'condition','grade',
          'yr_built','yr_renovated','padrao','custo-benefício',
          'lat', 'long']]

       
```

```python
#Criando dataframe para correlação 
df_corr = df[['id', 'city', 'price', 'price_ft²', 'bedrooms', 'bathrooms', 
          'sqft_lot', 'condition','grade',
          'yr_built','yr_renovated','padrao',
          'lat', 'long']]
    
    
```

```python
#Entendendo a correlação das variáveis
plt.figure(figsize=(15,15))
sns.heatmap(df_corr.corr(),annot = True, cmap='coolwarm')
```

```python
# Definindo as métricas de comparação
# De acordo com a mediana dos preços dos imóveis (dados centrais eliminando outlayers, um vez que no preço mediano encontram-se os imóveis com 
# com maior potencial de comercialização.#baseados na métrica condição maior que 3 e números de banheiros maiores que 2 
#baseados na métrica condição maior que 3 e números de banheiros maiores que 2 
# construídos após 1990
#ano de reforma     
df5_melhores = df2.loc[(df2['price'] > 321837) & (df2['price'] < 645000) &
                      (df2['condition'] > 3) & (df2['bedrooms'] > 2) &
                      (df2['grade'] > 8) & (df2['bathrooms'] > 2) &
                      (df2['yr_built'] > 1990)].sort_values(by="price", ascending = False).reset_index().head(5)
df5_melhores
```

```python
# Plotando um grafico do top 5 imoveis para reforma com a ID de cada um
fig = px.scatter_mapbox(df5_melhores, lat="lat", lon="long", hover_name="id",
                        hover_data=['price','bedrooms','bathrooms','condition','yr_built'],
                        zoom=3, height=500, size='price')

fig.update_layout(mapbox_style="open-street-map")
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})

fig.show()
```

```python
#filtrando os 5 imoveis recomendados
df5_piores = df2.loc[(df2['price'] > 645000) & (df2['price'] < 7700000) &
                      (df2['condition'] >3) & (df2['bedrooms'] < 3) &
                      (df2['bathrooms'] < 2) &
                      (df2['grade'] <7)].sort_values(by="price", ascending = False).reset_index().head(5)
df5_piores                     
```

```python
#Plotando os 5 imoveis não recomendados 
fig = px.scatter_mapbox(df5_piores  , lat="lat", lon="long", hover_name="id",
                        hover_data=['price','bedrooms','bathrooms','condition','yr_built'],
                        zoom=3, height=500, size='price')

fig.update_layout(mapbox_style="open-street-map")
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})

fig.show()
```
